```python
import plotly.express as px
from sklearn.datasets import load_iris
import pandas as pd

# Load the Iris dataset
iris = load_iris(as_frame=True)
df = iris.frame
df['species'] = iris.target_names[iris.target]

# Create the initial scatter plot
fig = px.scatter(
    df,
    x='sepal length (cm)',
    y='sepal width (cm)',
    color='species',
    size='petal length (cm)',
    hover_data=['petal width (cm)'],
    title='Interactive Iris Dataset Visualization'
)

# Add a dropdown menu to filter by species
fig.update_layout(
    updatemenus=[
        {
            'buttons': [
                {
                    'method': 'update',
                    'label': 'All Species',
                    'args': [{'visible': [True, True, True]}, {'title': 'All Species'}]
                },
                {
                    'method': 'update',
                    'label': 'Setosa',
                    'args': [{'visible': [True, False, False]}, {'title': 'Setosa Only'}]
                },
                {
                    'method': 'update',
                    'label': 'Versicolor',
                    'args': [{'visible': [False, True, False]}, {'title': 'Versicolor Only'}]
                },
                {
                    'method': 'update',
                    'label': 'Virginica',
                    'args': [{'visible': [False, False, True]}, {'title': 'Virginica Only'}]
                }
            ],
            'direction': 'down',
            'showactive': True,
        }
    ]
)

# Group the data by species to control visibility
fig.for_each_trace(lambda trace: trace.update(visible=True) if trace.name == 'setosa' else None)
fig.for_each_trace(lambda trace: trace.update(visible=True) if trace.name == 'versicolor' else None)
fig.for_each_trace(lambda trace: trace.update(visible=True) if trace.name == 'virginica' else None)

fig.show()
```

BUBBLE CHART

```python
import plotly.express as px
import pandas as pd

# Sample data
data = {
    'Wine Type': ['Type A', 'Type B', 'Type C', 'Type D', 'Type E'],
    'Alcohol': [10, 12, 9, 14, 11],       # X-axis
    'Sugar': [5, 7, 3, 8, 6],             # Y-axis
    'Quality': [3, 7, 4, 9, 5]            # Bubble size
}

df = pd.DataFrame(data)

# Create Bubble Chart
fig = px.scatter(df, 
                 x='Alcohol', 
                 y='Sugar', 
                 size='Quality',             # Bubble size
                 color='Wine Type',          # Bubble color
                 hover_name='Wine Type',     # Hover info
                 size_max=60,                # Max bubble size
                 title='Bubble Chart of Wine Types by Alcohol and Sugar Content')

fig.show()
```

Explanation of Parameters:

    x='Alcohol': Defines the variable on the x-axis.
    y='Sugar': Defines the variable on the y-axis.
    size='Quality': Controls the size of the bubbles based on wine quality.
    color='Wine Type': Differentiates bubbles by color for each wine type.
    hover_name='Wine Type': Shows wine type when you hover over a bubble.
    size_max=60: Controls the largest size the bubbles can grow.

BOX PLOT: A Box Plot is great for visualizing the distribution of numerical data and identifying outliers. It shows the median, quartiles, and potential outliers in a dataset.

```python
import plotly.express as px
import pandas as pd

# Sample data
data = {
    'Wine Type': ['Type A', 'Type B', 'Type A', 'Type B', 'Type A', 'Type B', 'Type A', 'Type B'],
    'Alcohol': [10, 12, 9, 14, 11, 13, 10.5, 12.5],
    'Quality': [5, 7, 4, 9, 6, 8, 5, 7]
}

df = pd.DataFrame(data)

# Create Box Plot
fig = px.box(df, 
             x='Wine Type', 
             y='Alcohol', 
             color='Wine Type',
             points='all',  # Shows all data points
             title='Alcohol Content by Wine Type')

fig.show()
```

Explanation of Parameters:

    x='Wine Type': Categorical variable on the x-axis.
    y='Alcohol': Numerical variable for which the distribution is visualized.
    color='Wine Type': Colors the boxes based on the wine type.
    points='all': Displays all individual data points along with the box plot.

Customizing the Box Plot: Display Only Outliers:

```python
fig = px.box(df, x='Wine Type', y='Alcohol', color='Wine Type', points='outliers',
             title='Box Plot Showing Only Outliers')
fig.show()
```

Horizontal Box Plot:

```python
fig = px.box(df, x='Alcohol', y='Wine Type', color='Wine Type',
             title='Horizontal Box Plot of Alcohol Content')
fig.show()
```

Add Category-Wise Comparison (e.g., Quality):

```python
fig = px.box(df, x='Quality', y='Alcohol', color='Quality',
             title='Alcohol Content by Wine Quality')
fig.show()
```

Customize Appearance:

```python
fig.update_traces(marker_color='purple', line_color='black', line_width=2)
fig.update_layout(title='Customized Box Plot')
fig.show()
```

A Histogram is used to visualize the distribution of a numerical variable by grouping data into bins. It helps in understanding the frequency distribution and identifying patterns like skewness, modality, and outliers.

```python
import plotly.express as px
import pandas as pd

# Sample data
data = {
    'Alcohol': [10, 12, 9, 14, 11, 13, 10.5, 12.5, 11.2, 13.8, 9.5, 12],
    'Wine Type': ['Type A', 'Type B', 'Type A', 'Type B', 'Type A', 'Type B', 
                  'Type A', 'Type B', 'Type A', 'Type B', 'Type A', 'Type B']
}

df = pd.DataFrame(data)

# Basic Histogram of Alcohol Content
fig = px.histogram(df, 
                   x='Alcohol', 
                   color='Wine Type',  # Different colors for each wine type
                   barmode='overlay',  # Overlapping bars
                   nbins=10,
                   title='Alcohol Distribution by Wine Type')

fig.show()
```

Explanation of Parameters:

    x='Alcohol': The numerical variable for the x-axis.
    nbins=10: Number of bins to group data.
    title='...': Title of the plot.
    barmode='overlay': Overlays the histograms for comparison.
    barmode='group': Places bars side by side for comparison.

An Area Chart is similar to a line chart but with the area below the line filled in, making it great for visualizing trends over time and comparing quantities across categories.

```python
import plotly.express as px
import pandas as pd

# Sample data
data = {
    'Year': [2015, 2016, 2017, 2018, 2019, 2020],
    'Wine Production': [1200, 1500, 1700, 1600, 1800, 1900],
    'Wine Type': ['White', 'White', 'White', 'White', 'White', 'White']
}

df = pd.DataFrame(data)

# Create Area Chart
fig = px.area(df, 
              x='Year', 
              y='Wine Production', 
              title='Wine Production Over Years')

fig.show()
```

Explanation of Parameters:

    x='Year': Variable on the x-axis (typically time).
    y='Wine Production': Numerical variable on the y-axis (quantities).
    title='...': Sets the title of the plot.

A Violin Plot combines aspects of a box plot and a kernel density plot to show the distribution of numerical data across different categories. Its particularly useful for visualizing the density of the data at different values, making it easier to spot multi-modal distributions (data with multiple peaks).

```python
import plotly.express as px
import pandas as pd

# Sample data
data = {
    'Wine Type': ['Type A', 'Type B', 'Type A', 'Type B', 'Type A', 'Type B', 'Type A', 'Type B'],
    'Alcohol': [10, 12, 9, 14, 11, 13, 10.5, 12.5],
    'Quality': [5, 7, 4, 9, 6, 8, 5, 7]
}

df = pd.DataFrame(data)

# Basic Violin Plot
fig = px.violin(df, 
                x='Wine Type', 
                y='Alcohol', 
                color='Wine Type', 
                title='Alcohol Content Distribution by Wine Type')

fig.show()
```

Explanation of Parameters:

    x='Wine Type': Categorical variable on the x-axis.
    y='Alcohol': Numerical variable for which the distribution is visualized.
    color='Wine Type': Colors the violins based on the wine type.

A Strip Plot (also known as a dot plot) displays individual data points along a categorical axis. Its great for visualizing the distribution of small datasets or showing all data points to detect clustering and outliers.

```python
import plotly.express as px
import pandas as pd

# Sample data
data = {
    'Wine Type': ['Type A', 'Type B', 'Type A', 'Type B', 'Type A', 'Type B', 'Type A', 'Type B'],
    'Alcohol': [10, 12, 9, 14, 11, 13, 10.5, 12.5],
    'Quality': [5, 7, 4, 9, 6, 8, 5, 7]
}

df = pd.DataFrame(data)

# Basic Strip Plot
fig = px.strip(df, 
               x='Wine Type', 
               y='Alcohol', 
               color='Wine Type', 
               title='Alcohol Content by Wine Type')

fig.show()
```

Explanation of Parameters:

    x='Wine Type': Categorical variable for the x-axis.
    y='Alcohol': Numerical variable for the y-axis.
    color='Wine Type': Colors the points based on wine type.
